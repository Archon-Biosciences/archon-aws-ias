```python
from Bio import SeqIO
```

```python
with open("../data/output.fasta") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        print(record.id)

        file_out = f"{record.id}.fasta"
        with open("../data/fasta/" + file_out, "w") as f_out:
            SeqIO.write(record, f_out, "fasta")
```
